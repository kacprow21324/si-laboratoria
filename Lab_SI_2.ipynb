```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
print(tf.__version__)
```

```python
training_data = np.load('TF_dataset.npz')
input_size = 2
output_size = 1
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model.compile(optimizer='lion', loss='MeanSquaredLogarithmicError')
model.fit(training_data['inputs'], training_data['targets'], epochs=50, verbose=1)
```

```python
model.layers[0].get_weights()
#targets = 20*xs + 25*zs - 8 + noise
```

```python
model.predict_on_batch(training_data['inputs']).round(1)
```

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])),
         np.squeeze(training_data['targets']), color='orange')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

```python
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model2.compile(optimizer='nadam', loss='MeanSquaredLogarithmicError')
model2.fit(training_data['inputs'], training_data['targets'], epochs=20, verbose=1)
```

```python
plt.plot(np.squeeze(model2.predict_on_batch(training_data['inputs'])),
         np.squeeze(training_data['targets']), color='red')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

```python
model3 = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model3.compile(optimizer='SGD', loss='huber')
model3.fit(training_data['inputs'], training_data['targets'], epochs=30, verbose=1)
```

```python
plt.plot(np.squeeze(model3.predict_on_batch(training_data['inputs'])),
         np.squeeze(training_data['targets']), color='green')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
